```python
# *: 변수 앞에 붙어 여러 개의 변수를 한번에 넣을 수 있는 컨테이너 역할
#    iterable객체에서의 unpacking 기능
# **: 딕셔너리를 unpacking

def print1(**kwargs):
  print(kwargs)
  print(type(kwargs))

print1(name = '권영주', age = 2)

# print()문에서의 " , "와 end 역할
print('I', 'love', 'you')         # 띄어쓰기
print('I' 'love' 'you')
for i in range(5):
  print(i, end='*')
```

```python
# default parameter
# hobby처럼 매개변수가 정해진 default parameter는 가장 끝에 위치해야 한다
def intro_me(name, age, hobby='노래듣기'):
  print(f"이름은 {name}입니다. ")
  print(f"나이는 {age}입니다. ")
  print(f"취미는 {hobby}입니다. ")
intro_me('권영주', 24, hobby='먹기')

# lambda 함수 (anonymous): 함수명이 없는 함수
add = lambda a, b: a + b
print(add(2, 3))

# colab or jupyter에서만 print() 없이 출력 가능
```

```python
# 변수 scope - 전역변수, 지역변수
# 전역변수는 되도록이면 건드리지 말 것
a = 1
def vartest(a):
  a += 1
  print(a)
  return a
b = vartest(a)
print(a, b)

gVar = 1
def vartest2():
  global gVar
  gVar += 1
vartest2()
gVar
```

```python
# 파일 file: 데이터를 보관하기 위한 장소 중 하나
# close()를 할 때까지 open()한 핸들러가 파일에 대한 권한을 쥐고 있음
f = open('매수종목.txt', 'w')
f.write('900260\n')
f.write('252670\n')
f.write('317240')
f.close()

stock = {'로스웰': 900260, 'KODEX 200선물인버스2X': 252670, 'TS트릴리온': 317240}
with open('매수종목1.txt', 'w') as s:     # with: 마지막에 close()를 자동 호출
  for k, v in stock.items():
    s.write(f"{k} {v}\n")
```

```python
lst = list()
f = open('매수종목.txt', 'r')
while True:
  line = f.readline()   # readline(): 한 줄씩 읽어옴, string
  if not line:
    break
  lst.append(line.strip())
f.close()
lst
```

```python
dic = dict()
with open('매수종목1.txt', 'r') as s:
  for line in s.readlines():     # readlines(): 파일 전 내용을 리스트로 반환
    line = line.strip()
    if len(line) <= 17:
      k, v = line.split()
    else:
      k1, k2, v = line.split()
      k = k1 + ' ' + k2
    dic[k] = v
dic
```
