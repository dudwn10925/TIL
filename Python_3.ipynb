```python
# *: 변수 앞에 붙어 여러 개의 변수를 한번에 넣을 수 있는 컨테이너 역할
#    iterable객체에서의 unpacking 기능
# **: 딕셔너리를 unpacking

def print1(**kwargs):
  print(kwargs)
  print(type(kwargs))

print1(name = '권영주', age = 2)

# print()문에서의 " , "와 end 역할
print('I', 'love', 'you')         # 띄어쓰기
print('I' 'love' 'you')
for i in range(5):
  print(i, end='*')
```

```python
# 변수 scope - 전역변수, 지역변수
# 전역변수는 되도록이면 건드리지 말 것
a = 1
def vartest(a):
  a += 1
  print(a)
  return a
b = vartest(a)
print(a, b)

gVar = 1
def vartest2():
  global gVar
  gVar += 1
vartest2()
gVar
```

```python
# 파일 file: 데이터를 보관하기 위한 장소 중 하나
# close()를 할 때까지 open()한 핸들러가 파일에 대한 권한을 쥐고 있음
f = open('매수종목.txt', 'w')
f.write('900260\n')
f.write('252670\n')
f.write('317240')
f.close()

stock = {'로스웰': 900260, 'KODEX 200선물인버스2X': 252670, 'TS트릴리온': 317240}
with open('매수종목1.txt', 'w') as s:     # with: 마지막에 close()를 자동 호출
  for k, v in stock.items():
    s.write(f"{k} {v}\n")
```

```python
lst = list()
f = open('매수종목.txt', 'r')
while True:
  line = f.readline()   # readline(): 한 줄씩 읽어옴, string
  if not line:
    break
  lst.append(line.strip())
f.close()
lst
```

```python
dic = dict()
with open('매수종목1.txt', 'r') as s:
  for line in s.readlines():     # readlines(): 파일 전 내용을 리스트로 반환
    line = line.strip()
    if len(line) <= 17:
      k, v = line.split()
    else:
      k1, k2, v = line.split()
      k = k1 + ' ' + k2
    dic[k] = v
dic
```

```python
# 클래스
# 메소드: 클래스 내의 함수(self 명시)
# __init__: 생성자 constructor, 객체 생성 시 자동 호출
# self: 생성된 인스턴스 지정
# self.: 객체 변수
# 클래스 변수: 클래스 내에 정의된 변수
# 클래스의 상속
# 모든 클래스는 object 클래스를 상속받음

class Baby:
  def __init__(self, **baby_info):
    self.baby_info = dict(baby_info)
    
  def getInfo(self):
    print(f"{self.baby_info['name']:>8}의 information")
    print('-' * 30)
    for key in self.baby_info:
      print(key, ': ', self.baby_info[key])
    print('\n')

  def getInfo_part(self, key):
    if key == 'name':
      print(f"{self.baby_info['name']}의 이름은 {self.baby_info[key]}입니다. \n")
    elif key == 'age':
      print(f"{self.baby_info['name']}의 나이는 {self.baby_info[key]}살 입니다. \n")
    elif key == 'sex':
      print(f"{self.baby_info['name']}의 성별은 {self.baby_info[key]}입니다. \n")
    elif key == 'hobby':
      print(f"{self.baby_info['name']}의 취미는 {self.baby_info[key]}입니다. \n")
    else:
      print('얻고자 하는 정보를 제대로 입력해주세요. \n')

  def setInfo(self, **baby_info):
    for key in baby_info:
      self.baby_info[key] = baby_info[key]

baby5 = Baby(name='권태우', age=18, sex='male')
baby3 = Baby(name='권영주', age=24, sex='female')
baby3.setInfo(hobby = 'sing a song')
baby3.getInfo()
baby3.getInfo_part('name')
baby5.getInfo_part('sex')
baby5.getInfo()
```

```python
class Car:
  def __init__(self, wheel, engine):
    self.wheel = wheel
    self.engine = engine

class Truck(Car):
  def __init__(self, wheel, engine, luggage):
    super().__init__(wheel, engine)
    self.luggage = luggage

truck = Truck(4, '2000cc', 2)
```

```python
# *: 가변 인수, packing or iterable object unpacking
# **: 키워드 가변 인수, dictionary packing or unpacking

def asterisk_test(a, *args):
  print(a, args)
  print(type(args))
asterisk_test(1, *(2, 3, 4, 5, 6))

def asterisk_test(a, **kwargs):
  print(a, kwargs)
  print(type(kwargs))
asterisk_test(1, b=2, c=3, d=4)
```
