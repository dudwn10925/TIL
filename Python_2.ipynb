```python
# List

# 리스트는 속도문제 때문에 같은 유형의 요소들만 모아두는 것이 좋음
# > lst = list('i am')
# > lst
# ['i', ' ', 'a', 'm']
comlist = [1, 2, ['a', 'b', ['life', 'is', 'good']]]
print(str(comlist[1]) + comlist[2][-1][1])    # 문자열과 숫자의 직접적인 덧셈 불가
print(comlist[-1][-1][1:])

a = [1, 2, 3, 4]
b = [7, 5, 6]
print(a + b)    # a.extend(b) => a = a + b, extend() arguments로 iterable 객체 가능
print(b * 2)
print(len(a))
del a[2:]
print(a)
a.append([3, 4])    # insert()는 append()에 비해 속도 저하
print(a)
b.sort()    # 문자 정렬 시 컴퓨터는 숫자만 인식하기에 아스키코드로 변환 후 정렬
print(b)
b.reverse()
print(b)
b.index(5)  # list에는 find()가 없음

print(a)
a.remove([3, 4])
print(a)
val = a.pop(1)
print(val)
```

```python
com_lan = ['C', 'Java']
inter_lan = ['Python']
com_lan.extend(inter_lan)
print(com_lan)
print(com_lan.count('Python'))
com_lan.extend('Python')
print(com_lan)
```

```python
nums = [1, 2, 3, 4, 5, 6]
print(f"max: {max(nums)}")
print(f"min: {min(nums)}")
print(f"sum: {sum(nums)}")
print('mean: {}'.format(sum(nums)/len(nums)))
print(nums[::2])
print(nums[1::2])
print(nums[::-1])   # nums.reverse(): 원본 데이터가 변함
```
