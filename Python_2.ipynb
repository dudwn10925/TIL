```python
# List

# 리스트는 속도문제 때문에 같은 유형의 요소들만 모아두는 것이 좋음
# > lst = list('i am')
# > lst
# ['i', ' ', 'a', 'm']
comlist = [1, 2, ['a', 'b', ['life', 'is', 'good']]]
print(str(comlist[1]) + comlist[2][-1][1])    # 문자열과 숫자의 직접적인 덧셈 불가
print(comlist[-1][-1][1:])

a = [1, 2, 3, 4]
b = [7, 5, 6]
print(a + b)    # a.extend(b) => a = a + b, extend() arguments로 iterable 객체 가능
print(b * 2)
print(len(a))
del a[2:]
print(a)
a.append([3, 4])    # insert()는 append()에 비해 속도 저하
print(a)
b.sort()    # 문자 정렬 시 컴퓨터는 숫자만 인식하기에 아스키코드로 변환 후 정렬
print(b)
b.reverse()
print(b)
b.index(5)  # list에는 find()가 없음

print(a)
a.remove([3, 4])
print(a)
val = a.pop(1)
print(val)
print(a)

# 삭제 시
# pop(index)
# del a(index)
# remove(element)
```

```python
# tuple
num_tuple = 1,    # =(1, )
print(num_tuple)
# tuple -> list, 강제 형변환 코드는 좋지 않음
electronics = tuple(['삼성전자', 'LG전자'])   # iterable 객체로 생성
print(electronics)
print(type(electronics))
electronics = list(electronics)
print(type(electronics))

fruits = ('apple', 'peach', 'watermelon')   # packing
a, p, w = fruits[:]    # unpacking
print(a, p, w)
```

```python
# dictionary

# 형식: {key1: value1, key2: value2}
# value 값으로 dictionary도 가능하다
dic1 = {1: '권영주0', 1: '권영주1', 'a': [1, 2, 3]}
dic2 = dict()
print(dic2)
# 추가
dic1["day"] = "20220111"
dic1['a', 'b', 'c'] = 123   # dictionary의 key값으로 list는 안 되지만 tuple은 가능
print(dic1)
# 삭제
del dic1[('a', 'b', 'c')]
print(dic1)
dic1.pop(1)   # pop(key)은 return 값이 존재
print(dic1)
# 검색
print(dic1['a'])
print(dic1.get('a'))
print('a' in dic1.keys())
print('b' in dic1)
print(dic1)
print(dic1.keys())
print(dic1.values())
print(dic1.items())
print(type(dic1.keys()))
print(type(dic1.values()))
print(type(dic1.items()))
for d in dic1.keys():
  print(f"{d}: {dic1[d]}")
dic1.clear()
print(dic1)
```

```python
# set 집합
set1 = set((1, 2, 3, 4))     # = set({1, 2, 3}) = set([1, 2, 3])
print(set1, type(set1))
set2 = {3, 4, 5, 6}
set1.add(5)
print(set1)
set1.remove(5)    # discard(): 없는 요소를 삭제 시 오류내지 않음
print(set1)
print(set2)
print(set1.intersection(set2))    # set1 & set2
print(set2.union(set1))           # set1 | set2
print(set2 - set1)
set1.update(set2)
print(set1)
```

```python
# Bool
print(type(2 > 1))
print(bool("python"))
print(bool(""))       # False: '', 0, None

print(id(set1))       # 메모리 상의 주소값
print(set1 is set2)   # 메모리 상의 주소값 비교
a, b = ['Python', 'Javascript']
print(a, b)
a, b = b, a
print(a, b)
```

```python
# 제어문

# if문
score = 80
result = 'pass' if score >= 80 else 'fail'    # if-else문을 한 줄로 작성 가능
print(result)

# while문
price_coffee = 4500
num_coffee = 3
while num_coffee > 0:
  print(f'커피 판매기\n남은 커피는 {num_coffee}잔 입니다. ')
  money = int(input('돈 입력: '))     # input()으로 받아들인 입력 값은 string
  if money > price_coffee:
    print(f'커피 나왔습니다.\n거스름돈 {int(money - price_coffee)}원 입니다.\n')
    num_coffee -= 1
  elif money == price_coffee:
    print("커피 나왔습니다.\n맛있게 드세요^^\n")
    num_coffee -= 1
  else:
    print('금액이 부족합니다.\n')

  if num_coffee == 0:
    print("판매 종료")
    break
```

```python
# for문
menu = ["뿌링클", "엽떡", '토실토실 방어', '후레쉬']
for m in menu:
  if m == "뿌링클":
    continue
  else:
    print(m)
print([m for m in menu[1:]])    # list comprehension

result = 0
for i in range(1, 10):
  result += i
print(result)
```

```python
# 함수
def add(*args):     # *args: args는 iterable객체이며 이 객체를 분해 가능
  print(args)
  result = 0
  for arg in args:
    print(arg)
    result += arg
  return result
print(add(1, 2, 3))
```
