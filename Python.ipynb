```python
print("I ate {0} apples. so I was sick for {day} days".format(100, day='seven'))
print("{0:*>7}".format('hi'))
print("{0:10.4f}".format(3.141592))
print("%10.4f" %3.141592)
```

```python
# f-formating
print(f"{'hi':$^10}")

# dictionary
d = {"name": '홍길동', 'age': 30}   # key - value
f"나의 이름은 {d['name']}입니다.\
 나이는 {d['age']}입니다."

print('{0}{1}{0}'.format('abra', 'cad'))
coord = {"latitude": '37.24N', 'longitude': "-115.81W"}
print('Coordinates: {latitude}, {longitude}'.format(**coord))    # **: 두 번 분해
f"Coordinates: {coord['latitude']}, {coord['longitude']}"
```

```python
# 문자열 함수
str = "Life is too short"
print(str.count('o'))
print(str.find('f'))    # 없는 문자로 find()하면 -1
print(str.index('f'))   # 없는 문자로 index()하면 error
print(', '.join('abcd'))
print(', '.join(['a', 'b', 'c', 'd']))

print(str.upper(), str.lower())
a = '  hi  '
print(a.lstrip())
print(a.rstrip())
print(a.strip())

print(str.replace('Life', 'Your leg'))    # 실질적으로 str이 변하는 것이 아님
a = 'a:b:c:d'
print(a.split(":"))
```
